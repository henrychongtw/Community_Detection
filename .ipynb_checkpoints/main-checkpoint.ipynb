```python
# Setting packages import
import os
import pandas as pd
import networkx as nx
import community
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = 'data/email-Eu-core.txt'

network_df = pd.read_csv(path, sep=" ", header=None)
network_df.columns = ['source', 'target']
```

```python
main_graph = nx.from_pandas_edgelist(network_df, 'source', 'target')
```

```python
path = 'data/question.txt'

q_df = pd.read_csv(path, sep=",", header=None)
q_df.columns = ['id','source', 'target']
```

```python
q_df.drop(q_df.index[0], inplace=True)
```

```python
q_df.drop('id', axis=1, inplace=True)
```

```python
q_df = q_df.reset_index(drop=True)
```

```python
q_graph = nx.from_pandas_edgelist(q_df, 'source', 'target')
```

```python
#first compute the best partition
partition = community.best_partition(main_graph)

```

```python
#drawing
community = {}

count = 0.
for com in set(partition.values()) :
    count = count + 1.
    list_nodes = [nodes for nodes in partition.keys()
                                if partition[nodes] == com]
    
    community[com] = list_nodes
```

```python
community
```

```python
def isSameCom(node1,node2):
    c1 = -1
    c2 = -1
    for key, value in community.items():
        if(int(node1) in value): 
            c1 = key
        if(int(node2) in value): 
            c2 = key
    
    if(c1 == c2): return 1
    else: return 0
```

```python
predict = []

predict = [isSameCom(q_df.loc[i, 'source'], q_df.loc[i, 'target']) for i in range(q_df.shape[0])]
```

```python
t_0 = 0
t_1 = 1
for i in predict:
    if(i == 0): t_0+=1
    else: t_1+=1
print(t_0,t_1)
        
```

```python
row = [i for i in range(len(predict))]
data = {'id': row, 'prediction': predict}
predict = pd.DataFrame(data=data, columns=['id', 'prediction'])
predict.to_csv("data/predict.csv", index=False)
print('Get predict file')
```
