```python
# community_train = sorted(map(sorted, nextlevel))

def create_dict(community_train):
    communities_dict = {}
    count = 0
    for comm in community_train:
        count += 1
        communities_dict[count] = comm
    return communities_dict

# communities = {}
communities = create_dict(communities_generate)
    
print(communities)
    
```

```python
def Check_Community(node1,node2):
    group1 = 0
    group2 = 0
    for key, value in communities.items():
        if(int(node1) in value):
            group1 = key
        if(int(node2) in value):
            group2 = key
            
    if(group1 == group2):
        return 1
    else:
        return 0
```

```python
predict = []
predict = [Check_Community(q_data.loc[i, "From"], q_data.loc[i,"To"]) for i in range(q_data.shape[0])]
```
